# BB84 Quantum Key Distribution — With Attacker (Eve)

This notebook simulates the BB84 protocol **with an attacker (Eve)** performing an **intercept-resend attack**, and demonstrates that the attack can be **detected**.

### Why the attack is detectable (Lecture 3b, slides 23–24)
The **No Cloning Theorem** prevents Eve from copying a qubit and re-sending the original.
Instead she must measure the qubit herself (destroying it) and re-prepare a new one based on her result.

When Eve guesses the **wrong basis**, she collapses the qubit into a random state.
Even when Alice and Bob later use **matching bases**, Bob may get the wrong bit — introducing errors.

Alice and Bob detect this by **comparing a subset of their sifted key bits** over a public channel (Lecture 3b, slide 24). A high error rate above a threshold indicates an attack.

### Parties
- **Alice** — encodes qubits
- **Eve** — intercepts, measures, and re-sends qubits
- **Bob** — measures the qubits Eve re-sends

```python
# ============================================================
# HELPER: Quantum Random Bit Generator
# ============================================================
# From Lecture 3b (slide 20): generate random bits by preparing
# |+> = H|0> and measuring. Gives 0 or 1 with probability 1/2.
# This is genuine quantum randomness, NOT Python's random library.

simulator = BasicSimulator()

def quantum_random_bit():
    """Generate a single random bit by measuring |+>."""
    qc = QuantumCircuit(1, 1)
    qc.h(0)          # Prepare |+> = H|0>
    qc.measure(0, 0)
    compiled = transpile(qc, simulator)
    result = simulator.run(compiled, shots=1).result()
    counts = result.get_counts()
    return int(list(counts.keys())[0])

def quantum_random_bits(n):
    """Generate a list of n random bits using quantum measurement."""
    return [quantum_random_bit() for _ in range(n)]

basis_label = lambda b: 's' if b == 0 else 'd'
print("Sample of 10 quantum random bits:", quantum_random_bits(10))
```

```python
# ============================================================
# ALICE — Encode qubits
# ============================================================
# From Lecture 3b (slide 16):
#   Standard basis (s): 0 -> |0>,  1 -> |1>
#   Diagonal basis (d): 0 -> |+>,  1 -> |->
# Diagonal encoding: apply H after setting the bit value.

NUM_QUBITS = 100   # Larger sample so error rate is statistically visible
ERROR_THRESHOLD = 0.1  # If >10% of check bits are wrong, report attack

alice_bits  = quantum_random_bits(NUM_QUBITS)
alice_bases = quantum_random_bits(NUM_QUBITS)  # 0=standard, 1=diagonal

def alice_encode(bit, basis):
    """
    Alice encodes one bit into a qubit.
    basis=0 (standard): 0->|0>, 1->|1>
    basis=1 (diagonal): 0->|+>, 1->|->
    """
    qc = QuantumCircuit(1)
    if bit == 1:
        qc.x(0)      # Set to |1>
    if basis == 1:
        qc.h(0)      # H|0>=|+>, H|1>=|->
    return qc

alice_qubits = [alice_encode(b, bas) for b, bas in zip(alice_bits, alice_bases)]

print("ALICE")
print("Bits  (first 20):", alice_bits[:20])
print("Bases (first 20):", [basis_label(b) for b in alice_bases[:20]])
```

```python
# ============================================================
# EVE — Intercept-Resend Attack
# ============================================================
# From Lecture 3b (slide 23):
#   The No Cloning Theorem prevents Eve from copying a qubit.
#   So Eve must measure Alice's qubit (destroying it) and
#   re-prepare a new qubit based on her measurement result,
#   then forward that to Bob.
#
#   Eve guesses a basis for each qubit (quantum random choice).
#   - If Eve guesses correctly -> she gets the right bit and
#     re-prepares the correct state -> Bob is unaffected.
#   - If Eve guesses wrong     -> she collapses the state randomly,
#     re-prepares wrong state  -> Bob may get wrong bit even when
#     his basis matches Alice's.
#
# This is the classic intercept-resend attack.

def eve_intercept(alice_qc, eve_basis):
    """
    Eve measures Alice's qubit in her chosen basis,
    then re-prepares and returns the qubit she will forward to Bob.
    """
    # Eve measures the qubit
    qc = alice_qc.copy()
    qc.add_register(__import__('qiskit').ClassicalRegister(1))
    if eve_basis == 1:          # Diagonal basis: H then measure
        qc.h(0)
    qc.measure(0, 0)
    compiled = transpile(qc, simulator)
    result = simulator.run(compiled, shots=1).result()
    counts = result.get_counts()
    eve_bit = int(list(counts.keys())[0])

    # Eve re-prepares a new qubit using her measured bit and basis
    # (she can't send the original — it was destroyed by measurement)
    new_qc = QuantumCircuit(1)
    if eve_bit == 1:
        new_qc.x(0)
    if eve_basis == 1:
        new_qc.h(0)
    return new_qc, eve_bit

eve_bases = quantum_random_bits(NUM_QUBITS)  # Eve's random basis choices

intercepted_qubits = []
eve_bits = []
for i in range(NUM_QUBITS):
    forwarded_qc, eve_bit = eve_intercept(alice_qubits[i], eve_bases[i])
    intercepted_qubits.append(forwarded_qc)
    eve_bits.append(eve_bit)

print("EVE (intercept-resend)")
print("Bases (first 20):", [basis_label(b) for b in eve_bases[:20]])
print("Bits  (first 20):", eve_bits[:20])
print(f"\nEve guessed Alice's basis correctly {sum(e==a for e,a in zip(eve_bases,alice_bases))} / {NUM_QUBITS} times")
```

```python
# ============================================================
# BOB — Measure the qubits forwarded by Eve
# ============================================================
# Bob is unaware of Eve. He measures each (intercepted) qubit
# in a randomly chosen basis, exactly as in the plain protocol.
# From Lecture 3b (slide 15): diagonal measurement = H then measure.

bob_bases = quantum_random_bits(NUM_QUBITS)

def bob_measure(qc, basis):
    """Bob measures a qubit in standard (basis=0) or diagonal (basis=1) basis."""
    qc_copy = qc.copy()
    qc_copy.add_register(__import__('qiskit').ClassicalRegister(1))
    if basis == 1:
        qc_copy.h(0)   # Diagonal measurement: H then measure
    qc_copy.measure(0, 0)
    compiled = transpile(qc_copy, simulator)
    result = simulator.run(compiled, shots=1).result()
    counts = result.get_counts()
    return int(list(counts.keys())[0])

bob_bits = [bob_measure(qc, bas) for qc, bas in zip(intercepted_qubits, bob_bases)]

print("BOB")
print("Bases (first 20):", [basis_label(b) for b in bob_bases[:20]])
print("Bits  (first 20):", bob_bits[:20])
```

```python
# ============================================================
# KEY SIFTING — Alice and Bob compare bases (public channel)
# ============================================================
# From Lecture 3b (slides 17-18):
#   Bob announces his bases publicly.
#   Alice announces which positions had matching bases.
#   They keep only those bits.

matching_indices = [i for i in range(NUM_QUBITS) if alice_bases[i] == bob_bases[i]]

alice_sifted = [alice_bits[i] for i in matching_indices]
bob_sifted   = [bob_bits[i]   for i in matching_indices]

print(f"Matching basis positions: {len(matching_indices)} / {NUM_QUBITS}")
print("Alice sifted (first 20):", alice_sifted[:20])
print("Bob   sifted (first 20):", bob_sifted[:20])
```

```python
# ============================================================
# ERROR CHECKING — Detect the attacker
# ============================================================
# From Lecture 3b (slide 24):
#   Alice and Bob compare a SUBSET of their sifted bits publicly.
#   Without an attacker: all compared bits should match.
#   With an attacker:    Eve's wrong-basis guesses introduce errors.
#
#   When Eve guesses the wrong basis (prob 1/2), she collapses
#   the qubit. Even when Alice and Bob both use the correct matching
#   basis, Bob gets a wrong bit with probability 1/2.
#   Overall error rate on matching-basis positions ~ 25%.
#
#   If the error rate exceeds ERROR_THRESHOLD, an attack is reported.
#   The bits used for checking are sacrificed (not used in final key).

# Use the first half of sifted bits for error checking
n_check = len(alice_sifted) // 2
check_alice = alice_sifted[:n_check]
check_bob   = bob_sifted[:n_check]

errors = sum(a != b for a, b in zip(check_alice, check_bob))
error_rate = errors / n_check if n_check > 0 else 0

print("--- ERROR CHECK (public comparison of subset) ---")
print(f"Check bits used  : {n_check}")
print(f"Errors found     : {errors}")
print(f"Error rate       : {error_rate:.1%}")
print(f"Threshold        : {ERROR_THRESHOLD:.1%}")
print()

if error_rate > ERROR_THRESHOLD:
    print("⚠️  ATTACK DETECTED! Error rate exceeds threshold.")
    print("   Alice and Bob should abort and discard the key.")
else:
    print("✅ No attack detected (error rate within threshold).")
    print("   (This is unlikely with Eve present — try again or increase NUM_QUBITS)")

# Final key = remaining sifted bits NOT used for error checking
alice_final_key = alice_sifted[n_check:]
bob_final_key   = bob_sifted[n_check:]

print()
print(f"Remaining key bits (if not aborted): {len(alice_final_key)}")
```

```python
# ============================================================
# SUMMARY — Why the error rate is ~25%
# ============================================================
# From Lecture 3b (slides 13-14, 23):
#
# When Eve intercepts a qubit:
#   - Prob 1/2: Eve picks the CORRECT basis.
#       She gets Alice's bit exactly, re-sends the correct state.
#       Bob's matching-basis measurement is undisturbed. Error prob = 0.
#   - Prob 1/2: Eve picks the WRONG basis.
#       She collapses |0>/|1> into |+>/|-> (or vice versa) randomly.
#       Her re-sent qubit is wrong. Even if Bob uses the correct basis,
#       he gets the wrong bit with prob 1/2.
#
# Expected error rate on matching positions = (1/2) * (1/2) = 25%
#
# Let's verify against our simulation:

total_matching = len(matching_indices)
total_errors_in_sifted = sum(a != b for a, b in zip(alice_sifted, bob_sifted))

print("--- THEORETICAL vs OBSERVED ---")
print(f"Expected error rate (theory) : ~25%")
print(f"Observed error rate (all sifted bits): {total_errors_in_sifted/total_matching:.1%}")
print(f"  (Over {total_matching} matching-basis bits)")
print()
print("Explanation:")
print("  Eve guesses wrong basis ~50% of the time.")
print("  When wrong, Bob gets the wrong bit ~50% of those times.")
print("  So ~25% of all matching-basis positions contain an error,")
print("  which Alice and Bob can detect by comparing a sample.")
print()
print("This confirms Lecture 3b slide 24:")
print("  'By sacrificing enough bits, the probability of detecting")
print("   the attack can be made arbitrarily high.'")
```
